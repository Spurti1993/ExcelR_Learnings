```python
#Vertical Stacking subplots in one directions or 1D
#1st way
plt.clf()
fig, axes = plt.subplots(2,figsize=(5,3))
axes[0].plot(x,y)
```

```python
#Vertical Stacking subplots in one directions or 1D
#2nd way
#this is applicable if we have less subplots to plot
plt.clf()
fig, (ax1,ax2) = plt.subplots(2,figsize=(3,2))
ax1.plot(x,y)
ax2.plot(y,x)
```

```python
#Horizontal Stacking subplots 
plt.clf()
fig, ax = plt.subplots(1,2,figsize=(5,3))#this will return both fig and axis obj
ax[0].plot(x,y)
ax[1].plot(y,x)
```

## Univariate → “one variable” data visualization

A strip plot is a graphical data anlysis technique for summarizing a univariate data set.
It is typically used for small data sets (histograms and density plots are typically preferred for larger data sets)
where one variable is categorical
A strip plot is simillar to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
#strip plot plots all numeric column only
#jitter - true measn it is scatter false means points will be plotted in a line by default jitter is true
sns.stripplot(data=tips,jitter=True)#tips is the dataset we are passing and it is mandatory parameter 
plt.ylabel('tip ($)')
plt.show()
sns.stripplot(y='tip',data=tips, jitter=False)#in y label what we pass
plt.show()
```

```python
#Grouping with stripplot()
#gives the idea of the density as it is scatter when jitter is true
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
#this kind of plot is called at dot plot and gives the idea of range of the data when jitter is false
sns.stripplot(x='day', y='tip', data=tips, size=8,jitter=False)#size Radius of the markers, in points.
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='smoker', y='total_bill',data=tips, jitter= 1)
#when jitter is given with value, as the jitter value increase te points are scatterd all over the plot less will give the proper estimation 
```

```python
#combining box plot and stripplot
sns.boxplot(data=tips)
sns.stripplot(data=tips)
```

```python
sns.stripplot(data=tips,x='day',y='tip',jitter=True, hue='smoker')#hue gives the differentiation btwn the categorical data
```

Swarmplot
Draw a categorical scatterplot with non-overlapping points.
This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. 
This style of plot is sometimes called a “beeswarm”.
this plot gives the count of each value plotted against without overlapping

```python
#Swarm plot
#gives ide of density
sns.swarmplot( data=tips, size=0.5)
plt.show()
sns.swarmplot(x='day', y='tip', data=tips, size=3)
plt.ylabel('tip ($)')
plt.show()
    
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')#hue means color i.e hue is used to diufferntiate the categorical data
plt.ylabel('tip ($)')
plt.show()
sns.swarmplot(x='day',y='tip',data=tips,hue='smoker')
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h',palette="seismic_r",size=3)#orient is the plcaing the plot in horizontal/vertical and palette is the color comnbination of the differentiation category that get seperated by hue
plt.xlabel('tip ($)')
plt.show()
```

```python
sns.swarmplot( x= 'time',y='total_bill',data=tips,order=["Lunch", "Dinner"])#order is nothing but considering the ordering value of the categorical data mentioned in the xaxis 
```

```python
#box and swarmplot
sns.boxplot(data=tips,y='tip',x='day')
sns.swarmplot(data=tips,y='tip',x='day',color='black',size=2)
```

## Bivariate → “two variables” data visualization

```python
#Joint plot
#Draw a plot of two variables with bivariate and univariate graphs.
#This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.
#plot with histogram and scatter plot mainly
sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
#kind is the plot type
#If True, remove observations that are missing from x and y
#smaller circle means more occurance /more dense in density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde',color='red',hue='day',palette='twilight_r',dropna=True)
plt.show()
```
