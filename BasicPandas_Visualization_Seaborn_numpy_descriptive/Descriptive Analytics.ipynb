# Missing values handling

```python
#autos = pd.read_csv('auto-mpg.csv')
#autos=pd.read_csv('auto-mpg.csv',header = None)#if header is present in the file then give it as none or else first row is considered as header which is wrong and intiger number to loadd the first row as header
#autos=pd.read_csv('auto-mpg.csv',header=0)
#autos = pd.read_csv('auto-mpg.csv',sep='\s+')#\s+ is the delimiter which is used to provide the index seperation of each coluymn when header is not given 

autos = pd.read_csv('auto-mpg.csv', sep ='\s+', header = None)
```

```python
autos.columns
```

```python
autos.columns = [ 'mpg','cylinders', 'displacement',
'horsepower', 'weight', 'acceleration',
'year', 'origin', 'name']
autos.head(5)
```

```python
autos.info()
```

```python
autos.horsepower.iloc[330:340]
```

```python
#errors{â€˜ignoreâ€™, â€˜raiseâ€™, â€˜coerceâ€™}, default â€˜raiseâ€™
#If â€˜raiseâ€™, then invalid parsing will raise an exception.
#If â€˜coerceâ€™, then invalid parsing will be set as NaN.
#If â€˜ignoreâ€™, then invalid parsing will return the input.
autos['horsepower'] = pd.to_numeric(autos['horsepower'],errors = "coerce") 
```

```python
#Find NaN values in the data
autos.horsepower.isnull().sum()
autos[autos.horsepower.isnull()]
```

```python
#Drop the NaN rows
autos = autos.dropna()
```

```python
#Check if any NaNs
autos[autos.horsepower.isnull()]
```
