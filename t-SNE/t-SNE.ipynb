```python

import pandas as pd
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster # mainluy for visualization
```

```python
# load data
filename = 'TSNE_data.csv'
dataframe = pd.read_csv(filename)
dataframe.head()
```

```python
# Split-out validation dataset
array = dataframe.values
# separate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
#TSNE visualization
data_tsne = TSNE(n_components=2).fit_transform(X)#n_components are no. of dimensions that we form max 3
cluster.tsneplot(score=data_tsne)# plts and downloads the data in the folder to show scatter plot without diffreentiation of clusters
```

```python
# get a list of categories
color_class = dataframe['diagnosis'].to_numpy()#used to return a NumPy ndarray representing the values in given Series or Index
#cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15, 1))#clsuterds are formed based on thr diagnosos
#use above or below
cluster.tsneplot(score=data_tsne, colorlist=dataframe['diagnosis'].values, legendpos='upper right',legendanchor=(1.15, 1))
#Plot will be stored in the default directory
```
